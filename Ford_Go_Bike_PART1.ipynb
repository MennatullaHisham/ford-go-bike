# Cleaning


```python
# create new dataframe copy
clean_df=df.copy()
```

## Define

```python
# Delete Unrequired columns
```

### Code

```python
clean_df.drop(columns=['start_station_longitude','start_station_latitude','end_station_latitude','end_station_longitude'],axis=1,inplace=True)
```

```python
clean_df
```

```python
clean_df.info()
```

## Duration in sec not easy for reading


### Define

```python
#Add minutes column, hours column and day column
```

```python
clean_df.insert(1,"duration_minutes",clean_df.duration_sec/60)
clean_df.insert(2,"duration_hours",clean_df.duration_sec/3600)
clean_df.insert(3,"duration_day",clean_df.duration_hours/24)
```

### Test

```python
clean_df.head()
```

```python
clean_df[clean_df.duration_hours>=23].count()[0]
```

## No "day" column to see the day name

### Define

```python
clean_df[['start_time','end_time']]=clean_df[['start_time','end_time']].apply(pd.to_datetime)
```

```python
clean_df.insert(4,"start_day",clean_df.start_time.dt.day_name())
clean_df.insert(6,"end_day",clean_df.end_time.dt.day_name())
```

### Test

```python
clean_df.info()
```

## Cleaning Quality issues

## Erroneous datatypes(start_time,end_time,start_station_id,nnend_station_id)

### Define

```python
# covert end_station_id,start_station_id to str as no operation will be performed on them
```

### Code

```python
clean_df[['end_station_id','start_station_id']]=clean_df[['end_station_id','start_station_id']].astype(str)
```

### Test

```python
clean_df.info()
```

```python
plt.figure(figsize=[14.70,8.27])
bins_edge=np.arange(0,45,1)
ticks=[0,5,10,20,30,40]
labels=['{}'.format(i) for i in ticks]
plt.hist(data=clean_df,x="duration_minutes",bins=bins_edge,rwidth=.8)
plt.title("Trip Duration in minutes",y=1.04,fontsize=14,fontweight='bold')
plt.xlabel('Duration in minutes' ,fontweight='bold')
plt.xticks(ticks,labels)
plt.ylabel('No. of bike trips' ,fontweight='bold');
```

### Insights
##### Average trip duration lies between 3 -> 14 min

```python
# what are the days where most trips
```

```python
clean_df.start_day.value_counts()
```

```python
plt.figure(figsize=[14.70,8.27])
colors=sb.color_palette()[-1]
s_day=clean_df.start_day.value_counts(ascending=True).index
sb.countplot(data=clean_df,x="start_day",color=colors,order=s_day)

plt.title("No. Of Bikes Starting EeveryDay",y=1.04,fontsize=14,fontweight='bold')
plt.ylabel('Days' ,fontweight='bold')
plt.xticks(rotation=45)
plt.xlabel('No. of bike trips' ,fontweight='bold');
```

### Insights
##### Most trips are on Thursday

## Bivariate Exploration

```python
plt.figure(figsize=[14.70,8.27])
sb.violinplot(data=clean_df.query('duration_minutes<60'),x='user_type',y='duration_minutes',color=colors)
plt.title("Duration of trips according to User type",fontweight='bold')
plt.xlabel('User Type',fontweight='bold')
plt.xlabel(' Duration Of Bike Trip',fontweight='bold');
```

### Insights
#### At duration less than or equal 60 min it seems like subscribers  achieves less than  customers

## Multivariate Exploration
